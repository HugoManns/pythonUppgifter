### Plotting in Matplotlib can be done in either of two ways, which ones? Which way is the recommended approach?

## Plotting in Matplotlib can be done in eiter of two ways:

- Object-oriented Approach:
This involves creating Figue and Axes objects explicitly and calling metods on them.
It is the recommended approcah because it provices more control and is better suited
for complex plots with multiple subplots or customization

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

fig, ax = plt.subplots()
ax.plot(x, y)
ax.set_title("Sin Wave")
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()
```

## The second way is Pyplot state based approach:
This uses the state-based interface provided by matplotlib.pyplot.
It is more simple and more concise but less flexible for complex plots.

```python
plt.plot(x, y)
plt.title("Sin Wave")
plt.xlabel("X")
plt.ylabel("Y")
plt.show()

# for most cases object-oriented is recommended
```

### Explain shortly what a figure, axes, axis and an artist is in Matplotlib.

```python
# 1. Figure:
# - Figures is a top-level container for all elements in a plot.
# - It acts as a blank slate where one or more plots are drawn.
# - You can think of it as the entire window or page where the visualization is created.

fig = plt.figure()
plt.show

# 2. Axes:
# - The axes is where the actual plot is created such as a scatter plot, line plot or bar plot.
# - It represents the plotting area within a Figure, including the data, lables, ticks and gridlines..
# - A Figure can contain multiple Axes also known as subplots.

fig, ax = plt.subplots()
ax.plot([1, 5, 10], [43, 44, 45], label="Line")
plt.show()


# 3. Axis:
# - An axis is part of an axes and refers to the x-axis or y-axis.
# - Both figures and Axes are atrists, as are elements like titles, lables, and gridlines for one dimensions of the plot.

fig, ax = plt.subplots()
ax.plot([1, 2, 3], [4, 5, 6])
ax.set_xlabel("X")
ax.set_ylabel("Y")
plt.show()

# 4. Artist:
# An artist is everything you see on the plot: lines, text, markers etc.
# Both figures and Axes are Artists, as are elements like titles, lables and legends.

fig, ax = plt.subplots()
line = ax.plot(x**2, label="Line")
ax.legend()
plt.show()

# In summary:
# - Figure: The overall container for the plot
# - Axes: The specific area where data is plotted (subplots)
# - Axis:  The individual x or y axis within an axes.
# - Artist: Any visual element in the plot
```

### When plotting in Matplotlib, what is the expected input data type?

```python
# Expected Input data types for plotting in matplotlib
#
# Matplotlib is flexible and can handle multiple data types for plotting. The most common ones are:
#
# 1. Lists or Tuples:
# - Python native data structures like lists or tuples are commonly used.
# Example:

x = [1, 2, 3, 4, 5]  # List
y = (10, 20, 30, 40, 50)  # Tuple
plt.plot(x, y)
plt.title("Using Lists and Tuples")
plt.show()

# 2. **NumPy Arrays**:
# - NumPy arrays are widely used for numerical computations and are ideal for plotting.
# - Example:
import numpy as np

x = np.linspace(0, 10, 100)  # NumPy array
y = np.sin(x)  # NumPy array
plt.plot(x, y)
plt.title("Using NumPy Arrays")
plt.show()

# 3. **Pandas Series**:
# - Matplotlib can directly plot Pandas Series for convenience.
# - Example:
import pandas as pd

data = pd.Series([5, 15, 25, 35, 45])
plt.plot(data)
plt.title("Using Pandas Series")
plt.show()

# 4. **Pandas DataFrame**:
# - When using DataFrames, you can specify columns to plot.
# - Example:
data = pd.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [10, 20, 30, 40, 50]
})
plt.plot(data["x"], data["y"])
plt.title("Using Pandas DataFrame")
plt.show()

# ### Summary:
# - Common input types include Python lists, tuples, NumPy arrays, Pandas Series, and DataFrames.
# - NumPy arrays and Pandas DataFrames are recommended for efficiency and compatibility with numerical operations.
```

### Create a plot of the function y = x^2 [from -4 to 4, hint use the np.linspace function] both in the object-oriented approach and the pyplot approach. Your plot should have a title and axis-labels.

```python
# Define the range and function
x = np.linspace(-4, 4, 100)  # Create 100 points from -4 to 4
y = x**2  # Compute y = x^2

# Object-Oriented Approach
fig, ax = plt.subplots()  # Create a Figure and Axes
ax.plot(x, y, label="y = x^2")  # Plot the function
ax.set_title("Object-Oriented Approach: y = x^2")  # Set title
ax.set_xlabel("X-axis")  # Set x-axis label
ax.set_ylabel("Y-axis")  # Set y-axis label
ax.legend()  # Add a legend
plt.show()

# Pyplot Approach
plt.plot(x, y, label="y = x^2")  # Plot the function
plt.title("Pyplot Approach: y = x^2")  # Set title
plt.xlabel("X-axis")  # Set x-axis label
plt.ylabel("Y-axis")  # Set y-axis label
plt.legend() # Add a legend
plt.show()
```

### Create a figure containing 2  subplots where the first is a scatter plot and the second is a bar plot. You have the data below. 

```python
# Data for scatter plot
np.random.seed(15)
random_data_x = np.random.randn(100)
random_data_y = np.random.randn(100)
x = np.linspace(-2, 2, 100)
y = x**2

# Data for bar plot
fruit_data = {'grapes': 22, 'apple': 8, 'orange': 15, 'lemon': 20, 'lime': 25}
names = list(fruit_data.keys())
values = list(fruit_data.values())
```

```python

# Create a figure with 2 subplots
fig, axes = plt.subplots(1, 2, figsize=(14, 5))  # 1 row, 2 columns

# Scatter plot (first subplot)
axes[0].scatter(random_data_x, random_data_y, alpha=0.5, label="Random Data")
axes[0].plot(x, y, color="red", label="y = x^2")
axes[0].set_title("Scatter Plot")
axes[0].set_xlabel("X-axis")
axes[0].set_ylabel("Y-axis")
axes[0].legend()

# Bar plot (second subplot)
axes[1].bar(names, values, color=['purple', 'red', 'orange', 'yellow', 'green'], label="Fruit Data")
axes[1].set_title("Bar Plot")
axes[1].set_xlabel("Fruit")
axes[1].set_ylabel("Quantity")
axes[1].legend()

# Adjust layout
plt.tight_layout()

# Show the plots
plt.show()
```
