```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
skodata =  pd.read_excel("skoStorlek_screentime.xlsx")
```

```python
print(skodata)
```

```python
plt.figure(figsize=(10, 6))
skodata.boxplot(column='Skärmanvändning', by='Skostorlek', grid=False)
plt.title('Skärmanvändning vs Skostorlek')
plt.suptitle('')
plt.xlabel('Skostorlek')
plt.ylabel('Skärmanvändning')
plt.tight_layout()
plt.show()
```

```python
x_min = skodata.groupby('Skostorlek')['Skärmanvändning'].min()
x_max = skodata.groupby('Skostorlek')['Skärmanvändning'].max()
x_mean = skodata.groupby('Skostorlek')['Skärmanvändning'].mean()

plt.figure(figsize=(10, 6))
plt.plot(x_min, color='red')
plt.plot(x_max, color='blue')
plt.plot(x_mean, color='green')
plt.title('Skostorlek by Skärmanvändning')
plt.xlabel('Skostorlek')
plt.ylabel('Skärmanvändning')
plt.tight_layout()
plt.show()
```
